```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def job_finder():
    key = input("Enter your skills for job search: ").strip()
    key_url = key.replace(" ", "%20")

    url = f"https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords={key_url}&txtLocation="
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "lxml")

    jobs = soup.find_all("li", class_="clearfix job-bx wht-shd-bx")

    # Lists to store job data
    company_names = []
    skills_list = []
    locations = []
    experiences = []

    for job in jobs:
        # Company Name
        company_tag = job.find("h3", class_="joblist-comp-name")
        company_name = company_tag.text.strip() if company_tag else "Not Available"
        company_name = " ".join(company_name.split())
        company_names.append(company_name)

        # Skills
        skills_div = job.find("span", class_="srp-skills")
        if skills_div:
            skills_text = skills_div.text.strip()
        else:
            skills_text = "Not Available"
        skills_list.append(skills_text)

        # Location
        location_tag = job.find("li", class_="location")
        if location_tag:
            location = location_tag.text.strip()
        else:
            location = "Not Available"
        locations.append(location)

        # Experience
        li_tags = job.find_all("li")
        exp_text = "Not Available"
        for li in li_tags:
            if li.find("i", class_="srp-icons experience"):
                exp_text = li.text.strip()
                break
        experiences.append(exp_text)

    # Build list of dictionaries (one per job)
    result = []
    for i in range(len(company_names)):
        job_data = {
            "Company Name": company_names[i],
            "Skills": skills_list[i],
            "Location": locations[i],
            "Experience": experiences[i],
        }
        result.append(job_data)

    print(f"\n Extracted {len(result)} job postings!")
    return result

# Run the job finder
jobs = job_finder()

# Create DataFrame and save to CSV
df = pd.DataFrame(jobs)
df.to_csv("scrapped_job_data.csv", index=False)

# Display DataFrame
df.head()
```
